```python
import os
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
dir = r'C:\Users\SHREY\Desktop\Task 1 Prodigy\train'
```

## Importing images from dataset and Reshaping

```python
categories = ['cat', 'dog']

data = []

for category in categories:
    path = os.path.join(dir, category)
    label = categories.index(category)
    
    for img in os.listdir(path):
        imgpath = os.path.join(path, img)
        pet_img = cv.imread(imgpath, 0)
        try:
            pet_img = cv.resize(pet_img, (100, 100), interpolation=cv.INTER_AREA)
        except:
            break
        image = np.array(pet_img).flatten()
        data.append([image,label])

print(len(data))
```

## Differentiating the Images(as NumPy array) and their corresponding Labels

```python
random.shuffle(data)

features = []
labels = []

for feature,label in data:
    features.append(feature)
    labels.append(label)
```

### Splitting the Dataset

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
```

# Training the model

```python
model = SVC(C=1, kernel='poly', gamma='auto') 
model.fit(x_train,y_train)
```

# Testing the model

```python
x = np.random.randint(1, high=800, size=10)

prediction = model.predict(x_test)
accuracy = model.score(x_test,y_test)
print('Accuracy :', accuracy)
for a in x:
    mypet = x_test[a].reshape(100,100)
    plt.imshow(mypet, cmap ='grey')
    plt.show()
    print('Prediction is', categories[prediction[a]])
```
